```python

import numpy as np

def generate_data(num_samples, matrix_size):
    # Generate random matrices A and B
    A = np.random.rand(num_samples, matrix_size, matrix_size)
    B = np.random.rand(num_samples, matrix_size, matrix_size)

    # Calculate the product of A and B
    C = np.array([np.dot(a, b) for a, b in zip(A, B)])

    return A, B, C

# Parameters
num_samples = 1000
matrix_size = 3

# Generate data
A_train, B_train, C_train = generate_data(num_samples, matrix_size)
A_test, B_test, C_test = generate_data(num_samples, matrix_size)
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Concatenate, Flatten

# Define the model architecture
input_shape = (matrix_size, matrix_size)

A_input = Input(shape=input_shape)
B_input = Input(shape=input_shape)

# Flatten the inputs
A_flattened = Flatten()(A_input)
B_flattened = Flatten()(B_input)

# Concatenate and Dense layers
merged = Concatenate()([A_flattened, B_flattened])
x = Dense(128, activation='relu')(merged)
x = Dense(matrix_size * matrix_size)(x)

# Reshape to matrix_size x matrix_size
output = tf.keras.layers.Reshape((matrix_size, matrix_size))(x)

model = Model(inputs=[A_input, B_input], outputs=output)
model.compile(optimizer='adam',
              loss='mean_squared_error')
# Train the model
model.fit([A_train, B_train], C_train,
          epochs=10,
          batch_size=32,
          validation_split=0.2)
# Evaluate the model
loss = model.evaluate([A_test, B_test], C_test)
print(f"Test Loss: {loss}")

# Make predictions
predictions = model.predict([A_test, B_test])

# Compare predictions with true results
for i in range(5):  # Display the first 5 examples
    print(f"Matrix A:\n{A_test[i]}")
    print(f"Matrix B:\n{B_test[i]}")
    print(f"True Product:\n{C_test[i]}")
    print(f"Predicted Product:\n{predictions[i]}")
    print("-----")
```

```python
import numpy as np

def is_prime(n):
    if n <= 1:
        return 0
    if n <= 3:
        return 1
    if n % 2 == 0 or n % 3 == 0:
        return 0
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return 0
        i += 6
    return 1

def generate_data(num_samples, max_value):
    X = np.random.randint(2, max_value, size=num_samples)
    y = np.array([is_prime(x) for x in X])
    return X, y

num_samples = 10000
max_value = 100

X_train, y_train = generate_data(num_samples, max_value)
X_test, y_test = generate_data(num_samples, max_value)
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input


model = Sequential([
    Input(shape=(1,)),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train,
          epochs=10,
          batch_size=32,
          validation_split=0.2)

loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy}")


predictions = model.predict(X_test)
predictions = (predictions > 0.5).astype(int)

for i in range(10):
    print(f"Number: {X_test[i]}, True Prime: {y_test[i]}, Predicted Prime: {predictions[i][0]}")
```
